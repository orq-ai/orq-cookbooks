## Create

Update or add a feedback of a log




Giving rating feedback:

```python
trace_id= "01JMMDA18Y2M86RSM0SM6B21X3" #paste a trace id of your log
```

```python
# Required parameters
field= "rating"
value= "good" # or "bad"

res = orq.feedback.create(field=field, value=value, trace_id=trace_id)
print (res)

print("")
print(f"Successfully created a {res.value} {res.property}.")
```

Giving correction feedback:

```python
# Required parameters
field= "correction"
value= "TEST user correction" # value of user correction

res = orq.feedback.create(field=field, value=value, trace_id=trace_id)
print (res)

print("")
print(f"Successfully created {res.property} with a value: {res.value}")
```
