## Create

Update or add user information to workspace




```python
# Required parameters
external_id = "12345"

# Optional parameters
display_name = "Test"
email = "user@example.com"
avatar_url = "https://media.istockphoto.com/id/1300845620/vector/user-icon-flat-isolated-on-white-background-user-symbol-vector-illustration.jpg?s=612x612&w=0&k=20&c=yBeyba0hUkh14_jgv1OKqIH0CCSWU_4ckRkAoy2p73o="
tags = ["3fa85f64-5717-4562-b3fc-2c963f66afa6"]
metadata = {"language": "eng"}

res = orq.contacts.create(external_id=external_id, display_name=display_name, email=email, avatar_url=avatar_url, tags=tags, metadata=metadata)
print (res)

print("")
print(f"Created contact {res.external_id} witha a name {res.display_name}")
```
